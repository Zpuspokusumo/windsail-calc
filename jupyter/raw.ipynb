```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load your data (replace 'your_data.csv' with your actual file)
data = pd.read_csv('saildata.csv')
print(data.head(5))
```

```python
# Extract features (angle) and target (lift coefficient)
X1 = data['Angle']
Y1 = data['Lift_Coefficient']
plt.scatter(X1, Y1)
plyf = np.polyfit(X1, Y1, 8)
f = np.poly1d(plyf)
print(f)
```

```python
X_new = np.arange(min(X1), max(X1))
Y_model = f(X_new)
plt.xlabel('Angle')
plt.ylabel('Lift Coefficient')
plt.scatter(X1, Y1)
plt.plot(X_new, Y_model)
```
