```python
import matplotlib.pyplot as plt
import numpy as np

# def generate_function():
#     def function(x):
#         return 1.406*(10**-8)*x**4 + -5.742*(10**-6)*x**3 + 0.000622*x**2 + -0.002353*x + -0.05202
#     return function

def generate_function():
    def function(x):
        return -1.234*(10**-15)*x**8 + 9.092*(10**-13)*x**7 - 2.726*(10**-10)*x**6 + 4.249*(10**-8)*x**5 - 3.621*(10**-6)*x**4 + 0.0001586*x**3 - 0.002737*x**2 + 0.01614*x - 0.0116
    return function

# Generate the function with the given coefficients
my_function = generate_function()

# Create a NumPy array of x values from 0 to 180
x_values = np.linspace(0, 180, 1000)  # 1000 points for smooth plot

# Calculate the corresponding y values using the function
y_values = my_function(x_values)

# Create the plot
plt.plot(x_values, y_values)

# Add labels and a title
plt.xlabel("x")
plt.ylabel("y")
plt.title("Plot of f(x)")

# Show the plot
plt.show()
```
